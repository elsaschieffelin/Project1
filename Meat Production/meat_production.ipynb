```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
f_path=os.path.join("meat_production.csv")
meat_production=pd.read_csv(f_path)
meat_production.head(100)
```

```python
wanted_columns = ["Country", "Year", "Livestock"]
meat_production = meat_production[wanted_columns]
meat_production
```

```python
meat_production.groupby("Country").count()
meat_production
```

```python
meat_production['Year'].astype(int)
```

```python
x = meat_production["Livestock"]
meat_production.plot.line("Year", "Livestock")
plt.title('Meat production')
plt.xlabel('Year')
plt.ylabel('Number of Livestock')
plt.grid()
```

```python
meat_production[meat_production["Country"] == "World"].plot.line("Year", "Livestock")

plt.title('Global Meat Production')
plt.xlabel('Year')
plt.ylabel('Livestock (Million Tonnes)')
plt.grid()

plt.savefig('Global Meat Production.png')
```

```python
meat_production[meat_production["Country"] == "United States"].plot.line("Year", "Livestock")

plt.title('United States Meat Production')
plt.xlabel('Year')
plt.ylabel('Livestock (Million Tonnes)')
plt.grid()

plt.savefig('United States Meat Production.png')
```

```python
x1 = meat_production[meat_production["Country"]== "Asia"]["Year"]
y1 = meat_production[meat_production["Country"] == "Asia"]["Livestock"]

x2 = meat_production[meat_production["Country"] == "Northern America"]["Year"]
y2 = meat_production[meat_production["Country"] == "Northern America"]["Livestock"]

x3 = meat_production[meat_production["Country"] == "Europe"]["Year"]
y3 = meat_production[meat_production["Country"] == "Europe"]["Livestock"]

x4 = meat_production[meat_production["Country"] == "South America"]["Year"]
y4 = meat_production[meat_production["Country"] == "South America"]["Livestock"]

x5 = meat_production[meat_production["Country"] == "Africa"]["Year"]
y5 = meat_production[meat_production["Country"] == "Africa"]["Livestock"]

x6 = meat_production[meat_production["Country"] == "Australia"]["Year"]
y6 = meat_production[meat_production["Country"] == "Australia"]["Livestock"]

fig=plt.figure()
ax=fig.add_subplot(111)
 
ax.plot(x1,y1,label='Asia')
ax.plot(x2,y2,label='Northern America')
ax.plot(x3,y3,label='Europe')
ax.plot(x4,y4,label='South America')
ax.plot(x5,y5,label='Africa')
ax.plot(x6,y6,label='Australia')


plt.title('Meat Production by Continent')
plt.xlabel('Year')
plt.ylabel('Livestock (Million Tonnes)')

plt.legend(loc= "left")
plt.grid()

plt.savefig('Meat Production by Continent.png')
```

```python
f_path=os.path.join("global-meat-production-by-livestock-type.csv")
livestock_type=pd.read_csv(f_path)
livestock_type.head(100)
```

```python
x1 = livestock_type[livestock_type["Country"] == "World"]["Year"]
y1 = livestock_type[livestock_type["Country"] == "World"]["Beef and Buffalo Meat"]

x2 = livestock_type[livestock_type["Country"] == "World"]["Year"]
y2 = livestock_type[livestock_type["Country"] == "World"]["Meat, Poultry"]

x3 = livestock_type[livestock_type["Country"] == "World"]["Year"]
y3 = livestock_type[livestock_type["Country"] == "World"]["Meat, pig"]

x4 = livestock_type[livestock_type["Country"] == "World"]["Year"]
y4 = livestock_type[livestock_type["Country"] == "World"]["Meat, duck"]

x5 = livestock_type[livestock_type["Country"] == "World"]["Year"]
y5 = livestock_type[livestock_type["Country"] == "World"]["Meat, game"]

x6 = livestock_type[livestock_type["Country"] == "World"]["Year"]
y6 = livestock_type[livestock_type["Country"] == "World"]["Sheep and Goat Meat"]



fig=plt.figure()
ax=fig.add_subplot(111)
 
ax.plot(x1,y1,label='Beef and Buffalo Meat')
ax.plot(x2,y2,label='Meat, Poultry')
ax.plot(x3,y3,label='Meat, pig')
ax.plot(x4,y4,label='Meat, duck')
ax.plot(x5,y5,label='Meat, game')
ax.plot(x6,y6,label='Sheep and Goat Meat')


plt.title('Global Meat Production By Livestock Type')
plt.xlabel('Year')
plt.ylabel('Livestock Type (Million Tonnes)')

plt.legend(loc="left")
plt.grid()

plt.savefig('Global Meat Production By Livestock Type.png')
```
